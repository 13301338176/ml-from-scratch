# Monte Carlo methods

- Estimating the value of `pi`

```python
import numpy as np
import matplotlib.pyplot as plt

N = 1000
M = 0

X, Y = [], []
Z = []
E = []

np.random.seed(100)

for i in range(1, N+1):
    x, y = np.random.random(2)
    
    if pow(x, 2) + pow(y, 2) <= 1.0:
        M+=1
        label = 1
    else:
        label = 0
        
    estimate = M/i * 4
    error = np.pi - estimate
    
    X.append((x, y))
    Y.append(label)
    Z.append(estimate)
    E.append(error)
        
    if i % 10000 == 0:
        print("Estimation at {:>8d}: {}".format(i, estimate)) 

print("Final estimate at {}: {}".format(N, M/N * 4))

X = np.asarray(X)
Y = np.asarray(Y)
Z = np.asarray(Z)
E = np.asarray(E)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.show()
```

```python
plt.plot(np.arange(N), Z)
plt.plot(np.arange(N), E)
```
